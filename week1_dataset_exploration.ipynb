```python
# Map dataset item to your pipeline format
def dataset_to_pipeline_format(item):
    return {
        "category": item["articleType"],
        "color": item["baseColour"],
        "gender": item["gender"],
        "season": item["season"],
        "description": item["productDisplayName"],
        "attributes": [item["usage"]]
    }

sample = dataset_to_pipeline_format(dataset[0])
print(sample)
```

```python
from fashion_data import process_fashion_dataset

# Convert 10 real items to pipeline format
raw_items = [dataset_to_pipeline_format(dataset[i]) for i in range(10)]

# Clean them
cleaned_items = process_fashion_dataset(raw_items)

# Print summary
for item in cleaned_items:
    print(f"{item['category']} | {item['color']} | {item['gender']} | {item['season']}")
```
